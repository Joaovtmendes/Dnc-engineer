```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
```

```python
alunos = pd.read_excel("aprovacao_alunos.xlsx")
alunos.head()
```

```python
alunos.describe()
```

```python
alunos.info()
```

```python
scaler= MinMaxScaler()

alunosN = pd.DataFrame(scaler.fit_transform(alunos), index=alunos.index, columns =alunos.columns)

alunosN.describe()
```

```python
sns.regplot(x="ano_2015",y="ano_2016", data=alunosN)
```

```python
import statsmodels.formula.api as smf
function = 	"ano_2015~ano_2016-1"
model = smf.ols(formula=function, data=alunosN).fit()
print(model.summary())
```

```python
x = alunosN[["ano_2015"]]
y = alunosN[["ano_2016"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

lr = LinearRegression()

lr.fit(x_train,y_train)

LinearRegression();
```

```python
r_sq= lr.score(x_train,y_train)
print(r_sq)
```

```python
y_pred_train = lr.predict(x_train)
print("MAE:", metrics.mean_absolute_error(y_train,y_pred_train))
print("MSE:", metrics.mean_squared_error(y_train,y_pred_train))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_train,y_pred_train)))
```

```python
lr.fit(x_test,y_test)
LinearRegression();
```

```python
y_pred_test = lr.predict(x_test)
print("MAE:", metrics.mean_absolute_error(y_test,y_pred_test))
print("MSE:", metrics.mean_squared_error(y_test,y_pred_test))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test,y_pred_test)))
```
